```python
from tensorflow.keras.models import model_from_json
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import numpy as np
```

# Loading model

```python
json_file = open('model.json','r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
loaded_model.load_weights('model.h5')
```

# Prediction

```python
img = mpimg.imread(r'C:\Users\prasanna\Desktop\SEM-2\tensorflow-animals\tf_files\animals\dog\pic_004.jpg')
plt.imshow(img)
```

```python
img = cv2.imread(r'C:\Users\prasanna\Desktop\SEM-2\tensorflow-animals\tf_files\animals\dog\pic_004.jpg',cv2.IMREAD_GRAYSCALE)
new_img = cv2.resize(img,(60,60))
new_img = new_img.reshape(-1,60,60,1)
predict = loaded_model.predict(new_img)
np.argmax(predict)
categories = ['butterfly','cat','cock','cow','dog','elephant','horse','sheep','spider','squirrel']
categories[np.argmax(predict)]
```
